# 20.classifie movie based on rating(excellent good and average)

```python
def rating(rating):
    if rating>7:
        return "Excellent"
    elif rating>=6:
        return "Good"
    else:
        return "Average"

        
```

# 22. Find unique value from the Genre

```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique_list=[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
```

```python
unique_list
```

```python
len(unique_list)
```
